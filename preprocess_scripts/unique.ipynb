```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../2023MCM_code/unigram_freq.csv')
df
```

```python
data = pd.read_excel('datafeatures_all.xlsx')
```

```python
data.insert(loc=len(data.columns), column='appear freq', value=np.zeros(len(data)))
data
```

```python
for i in range(len(data)):
    # print(data['Word'][i], df.loc[df['word'] == data['Word'][i]]['count'].values)
    # data['appear freq'][i] = df.loc[df['word'] == data['Word'][i]]['count'].values
    data['appear freq'].iloc[i] = int(df.loc[df['word'] == data['Word'][i]]['count'].values)
```

```python
data['appear freq'] = np.log2(data['appear freq'])
data
# - data['appear freq'].mean(axis=0)) / data['appear freq'].std(axis=0)
```

```python
data['appear freq'] = (data['appear freq'] - data['appear freq'].mean(axis=0)) \
        / data['appear freq'].std(axis=0)
data
```

```python
data['appear freq'] = (data['appear freq'] - data['appear freq'].mean(axis=0)) / data['appear freq'].std(axis=0)
print(np.corrcoef(data['appear freq'], data['1 try']))
print(np.corrcoef(data['appear freq'], data['2 tries']))
print(np.corrcoef(data['appear freq'], data['3 tries']))
print(np.corrcoef(data['appear freq'], data['4 tries']))
print(np.corrcoef(data['appear freq'], data['5 tries']))
print(np.corrcoef(data['appear freq'], data['6 tries']))
print(np.corrcoef(data['appear freq'], data['num_phonemes']))
```
